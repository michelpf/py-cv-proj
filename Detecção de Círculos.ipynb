```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('imagens/circles.png')
image_original = cv2.imread('imagens/circles.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

blur = cv2.medianBlur(gray, 3)

plt.figure(figsize=(8, 4)) 
plt.imshow(blur)
plt.suptitle('Imagem original em grayscale')
plt.show()

circles = cv2.HoughCircles(blur, cv2.HOUGH_GRADIENT, 1, 45, param1=190, param2=14.5, minRadius=5)

i_circle = 0
for i in circles[0,:]:
    i_circle+=1
    radius = i[2];
    print("Detectado círculo " + str(i_circle) + ", diâmetro " + str(2*radius))
    cv2.circle(image,(i[0], i[1]), i[2], (255, 0, 0), 2)

plt.figure(figsize=(8, 4)) 
plt.imshow(image)
plt.suptitle('Círculos detectados ' + str(i_circle))
plt.show()

f = plt.figure(figsize=(16, 4))
f.add_subplot(1,2, 1)
plt.imshow(image_original)
f.add_subplot(1,2, 2)
plt.imshow(image)
plt.suptitle('Comparação (Círculos detectados ' + str(i_circle)+ ')')
plt.show(block=True)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image_original = cv2.imread('imagens/shapes_leo.jpg')
image = cv2.imread('imagens/shapes_leo.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

kernel = np.ones((4, 4), np.uint8)

erosao = cv2.erode(gray, kernel, iterations = 2)
dilatacao = cv2.dilate(erosao, kernel, iterations = 2)

blur = cv2.medianBlur(erosao, 3)

plt.figure(figsize=(8, 4)) 
plt.imshow(blur)
plt.suptitle('Imagem original em grayscale')
plt.show()

circles = cv2.HoughCircles(blur, cv2.HOUGH_GRADIENT, 1, 45, param1=190, param2=40, minRadius=5)

i_circle = 0
for i in circles[0,:]:
    i_circle+=1
    radius = i[2];
    print("Detectado círculo " + str(i_circle) + ", diâmetro " + str(2*radius))
    cv2.circle(image,(i[0], i[1]), i[2], (255, 0, 0), 2)

plt.figure(figsize=(8, 4)) 
plt.imshow(image)
plt.suptitle('Círculos detectados ' + str(i_circle))
plt.show()

f = plt.figure(figsize=(16, 4))
f.add_subplot(1,2, 1)
plt.imshow(image_original)
f.add_subplot(1,2, 2)
plt.imshow(image)
plt.suptitle('Comparação (Círculos detectados ' + str(i_circle)+ ')')
plt.show(block=True)
```
